
## Tokenization Using Python’s split() Function
This is the simplest form of tokenization.
It splits text wherever a space character appears.
While fast and easy, it doesn’t handle punctuation, emojis, or contractions.


```python
# Basic whitespace-based tokenization
text = """Natural Language Processing (NLP) is a subfield of AI
that helps computers understand human language."""

tokens = text.split()  # Splits wherever there is whitespace
print("Tokens:", tokens)
print("Number of tokens:", len(tokens))
```

## Tokenization Using Regular Expressions (RegEx)
Regular expressions let us control exactly what gets captured as a token.
Here, we use a pattern that extracts only alphabetic words, skipping punctuation and numbers.


```python
import re   # Regular expressions library

text = """Natural language processing (NLP) is a subfield of linguistics,
computer science, and AI."""

# Extract alphabetic sequences only
tokens = re.findall(r"[A-Za-z]+", text)

print("Tokens:", tokens)
```

## Tokenization Using NLTK
NLTK provides robust tokenizers trained on linguistic data.
We’ll demonstrate both word-level and sentence-level tokenization.
If you see a LookupError for 'punkt_tab', install it using nltk.download('punkt_tab').


```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

# Download both required tokenizer packages
nltk.download('punkt')
nltk.download('punkt_tab')   # <-- Added line to fix LookupError in newer NLTK versions

text = "NLP makes machines understand human language. It’s fascinating!"

# Word-level tokenization
word_tokens = word_tokenize(text)

# Sentence-level tokenization
sentence_tokens = sent_tokenize(text)

print("Word Tokens:", word_tokens)
print("Sentence Tokens:", sentence_tokens)
```

## Tokenization Using spaCy
spaCy is an advanced NLP library that tokenizes and also provides linguistic features.
Each token carries attributes like lemma, part of speech, and syntactic relationships.

```python
import spacy

nlp = spacy.load("en_core_web_sm")  # Load English pipeline
text = "SpaCy makes NLP tasks simple, fast, and efficient."

doc = nlp(text)  # Process the text through spaCy

tokens = [token.text for token in doc]
print("Tokens:", tokens)

# Display token attributes for illustration
print("\nToken | Lemma | POS")
print("-" * 40)
for token in doc:
    print(f"{token.text:15} {token.lemma_:15} {token.pos_}")
```

## Tokenization Using Keras
Keras includes a simple preprocessing utility for word tokenization.
It automatically lowercases and removes punctuation—ideal for neural network preprocessing.


```python
# import path for TensorFlow versions
from tensorflow.keras.preprocessing.text import text_to_word_sequence

text = """Natural Language Processing allows computers to
analyze large amounts of text."""

tokens = text_to_word_sequence(text)
print("Tokens:", tokens)
```

## Visualizing Token Frequencies
Visualization helps students see which words occur most often after tokenization.
Here, we’ll plot a simple word frequency bar chart.


```python
from collections import Counter
import matplotlib.pyplot as plt

text = "NLP is fun and NLP is powerful because NLP learns patterns."
tokens = text.lower().split()

freq = Counter(tokens)

plt.figure(figsize=(6,3))
plt.bar(freq.keys(), freq.values(), color='skyblue')
plt.title("Word Frequency Distribution")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.show()
```

## Subword Tokenization Using Hugging Face
Modern transformer models use subword tokenization to handle unseen words.
This example shows how a pretrained BERT tokenizer breaks a word into smaller pieces.



```python
# Install transformers if not available
!pip install transformers --quiet

from transformers import AutoTokenizer

# Load the pretrained BERT tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Use deliberately complex or uncommon words
text = "unhappiness reconstruction microlearning artificiality playfulness"

tokens = tokenizer.tokenize(text)

print("Subword Tokens:", tokens)
```

### Understanding the “##” Symbols in Subword Tokenization

When we use subword tokenizers like **BERT’s WordPiece**, each token is either:
- a **standalone word**, or  
- a **continuation of another word**.

Tokens that start with `##` indicate that they are **subwords** (continuations)
and should be attached to the previous token when reconstructing the word.

Example:  
- `unhappiness` → `['un', '##happiness']`  
  → The model reads this as *"un" + "happiness"*.  
- `reconstruction` → `['re', '##construction']`  
  → The prefix `re` and root `construction` are learned separately.

This system helps the model handle:
- **rare or unseen words** by combining known sub-tokens.  
- **vocabulary efficiency**, since the model only stores common prefixes, suffixes, and roots.

So, whenever you see a token beginning with `##`,  
it means *“this piece continues from the previous token.”*
